```python
import pandas as pd
import re
from selenium import webdriver
```

```python
# init driver
driver = webdriver.Chrome()
base_url = 'https://www.kooora.com/?n=0&o=n&pg={}'

data = pd.DataFrame({
    'title': [], 'body': []
})

regexp = re.compile(r'www.kooora.com')


def get_articles_page(url):
    driver.get(url)
    page = driver.find_element_by_class_name('newsList')
    return page


def get_articles_list(page):
    links = []
    article_links = page.find_element_by_tag_name(
        'ul').find_elements_by_xpath('//li[@data-date]/div[@class="inline"]')
    # print(article_links)
    for item in article_links:
        link= item.find_element_by_tag_name('a')
        url = link.get_attribute("href")
        # print(url)
        links.append(url)

    #     if regexp.search(url):
    return links


def scrape_article(data, article):
    driver.get(article)
    title = driver.find_element_by_id('articleTitle').text
    body = driver.find_element_by_class_name('articleBody').text
    row = {'title': title, 'body': body}
    data = data.append(row, ignore_index=True)
    return data

for page_num in range(23500, 24000):
    url = base_url.format(page_num)
    page = get_articles_page(url)
    articles = get_articles_list(page)
    for article in articles:
        # compplete scrape article
        # print(article)
        data = scrape_article(data, article)


```
